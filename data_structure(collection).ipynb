```python
# Defination
# List is a data structure used to store multiple
# types of data in one place.

# Properties
# 1. List are mutable(update) in nature.
# 2. List can be hetrogenous and homogeneous.
# 3. List supports indexing,slicing and in ordered.
# 4. List allow duplicate value to store.
```

```python
# Creation.
# 1.method using []
stu_marks=[56,44,78,33,89]
print(stu_marks[0])

# 2.method using list() 
stu_name=list(["aman","rahul","sony"])
print(stu_name)

stu_details=[89,"rahul",89,"aman",True,False]
print(stu_details)
```

```python
# Indexing
stu_marks=[56,44,["aman","rahul",[100,200]],78,33,89]
print(stu_marks[-3])
```

```python
# update 
stu_marks=[56,44,78,33,89,["aman","sony"]]
stu_marks[0]=500
stu_marks[5][0]="Dev"
print(stu_marks)
```

```python
# Slicing  [start-0:end-1:step-1]
stu_marks=[10,20,30,40,50,60,70,80,90]
sliced=stu_marks[::-1]
print(sliced)
```

```python
stu_marks=[10,20,30,40,50,60,70,80,90]
total=0
size=len(stu_marks)
for i in range(size):
    total=total+stu_marks[i]
print(total)
```

```python
# in-built methods
# append()
stu_marks=[50,56,89,88]
# stu_marks.append(["Dev","aman"])
stu_marks.extend([["Dev","aman"]])
print(stu_marks)
```

```python
# extend()
stu_marks=[50,56,89,88]
stu_marks.extend([[5,6,7,7,700],["a","b"],[67,90]])
print(stu_marks[6][0])
```

```python
# insert()
stu_marks=[50,56,89,88]
stu_marks.insert(2,"aman")
print(stu_marks)
```

```python
# pop() : by default remove from last but also works with index.
stu_marks=[50,56,89,88]
res=stu_marks.pop(0)
# stu_marks.pop()
print(res)
print(stu_marks)
```

```python
# remove()
stu_marks=[50,56,89,88]
stu_marks.remove(56)
print(stu_marks)
```

```python
# reverse()
stu_marks=[50,56,89,88,"a","b"]
stu_marks.reverse()
print(stu_marks)
```

```python
# sort() 
stu_marks=[50,56,89,88]
stu_marks.sort(reverse=True)
print(stu_marks)
```

```python
# count() 
stu_marks=[50,56,50,89,88,50]
res=stu_marks.count(50)
print(res)
```

```python
# index() 
stu_marks=[50,56,50,89,88,50]
res=stu_marks.index(56)
print(res)
```

```python
# copy()
stu_marks=[50,56,50,89,88,50]
new_stu_marks=stu_marks.copy()
new_stu_marks.pop()
print(new_stu_marks)
print(stu_marks)
```

```python
# clear()
stu_marks=[50,56,50,89,88,50]
stu_marks.clear()
print(stu_marks)
```

```python
stu_marks=[10,20,30,40,50,60,7,80,9]
graced_marks=[str(i)+":Even" if i%2==0 
              else str(i)+":Odd"
              for i in stu_marks]
print(graced_marks)
```
