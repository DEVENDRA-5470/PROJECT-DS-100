```python
# wap to store and print the sum of only even numbers
# and in the same program print the product of odd numbers.
s=0
m=1
for i in range(1,5):
    if i%2==0:
        s=s+i
    else:
        m=m*i
print(s,m)
```

```python
# wap to count all even and odd numbers.
num=12
even=0
odd=0
for i in range(1,num+1): 
    if i%2==0:
        print(i)
        even+=1
    else:
        odd+=1
print("Sum of even",even)
```

```python
text="Lorem"
# 1.Find what is the frequency of 'm'.
# 2.Find the count of consonants.
count=0
cons_count=0
for i in text:
    if i not in "aeiouAEIOU":
        cons_count+=1
        
    if i=="m" or i=="M":
        count+=1
print(count)
print(cons_count)
```

```python
# count_frequency of how many times same char appears.
text="this is text"
target="t"
count=0
for i in range(len(text)):
    for j in range(1):
        if text[i] == target:
            count+=1
print(f"Count of {target} : {count}")
```

```python
# square
n=4
for i in range(n):
    for j in range(n):
        if i%2==0:
            print(j,end=" ")
        else:
            print(i,end=" ")
    print()
    
```
