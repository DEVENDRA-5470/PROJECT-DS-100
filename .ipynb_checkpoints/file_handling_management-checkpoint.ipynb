```python
# What is file handling.
# File handling is a of creating,opening,reading
# writing ,updating and closing files.
# so a program can store data permanently.

# What is file 
A File is a structured collection of data stroed
on a storage device(SSD,Hard drive etc) with 
a specific name and location managed by OS.

# Types of file.
# Text files (.csv,.docs,.txt)
# Binary file (media files.)

# Path and its types.
# Path is the address of a file or directory in  a file system.
# ex.c:/user/dev/python.py

# 1.Absolute Path
    An absolute path is the full path from the root directory.
    C:\
    D:\
    E:\
# 2.Relative Path
    A Relative path depends on you current working directory (CWD).
    BATCH-100/text.py
    downloads/resume.pdf
    .  current dir
    .. parent dir

# Methods.
# open("file_name","mode",encode=) : Used to open file.
# read()
# readlines()
# write()
# writelines()
# close()
# flush()
# seek()
# tail()


# Modes:
x  : only file creates
w  : file creates and write
w+ : file write and read
r  : file read
r+ : open and read (cursor start from 0)
a  : append data in existing file.

# Binary Modes
rb,wb,ab.

# Operations
# CRUD operation
# Context manager
```

```python
# Creating a file.
import os
file_name="data.txt"
if os.path.exists(file_name): # It checks existense of full path.
    print("File already hai ..")
else:
    file=open(file_name,"x")
    print("File Created")
```

```python
# Writing data in a file.
file_name="data.txt"
file=open(file_name,"w")
file.write("This data from file handling")
file.close()
print("File written")
```

```python
file_name="dev.txt"
file=open(file_name,"w")
file.write("This dev from file handling")
file.flush()
file.close()
print("File written")
```

```python
with open("new.txt","w+",encoding="utf-8") as file:
    file.write("This is Done in one time ❌✅ ✅✅✅")
    file.seek(3)
    read=file.read(99)
    print(read)
    print("File done")
```

```python
from faker import Faker
fake=Faker()

with open("fake.txt","w+") as file:
    for i in range(100000):
        file.write(fake.name()+"\n"+fake.email()+"\n"+"-----------\n")
print("Done")
```

```python
# update a file.
with open("data.txt","w+") as file:
    file.write("This is first line of the file")
print("File Written")
```
