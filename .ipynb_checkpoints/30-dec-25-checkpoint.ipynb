```python
import time
import pyautogui
from datetime import datetime
from pathlib import Path

out = Path("screens")
out.mkdir(exist_ok=True)

while True:
    ts = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    img = pyautogui.screenshot()
    img.save(out / f"screen_{ts}.png")
    time.sleep(30)
```
