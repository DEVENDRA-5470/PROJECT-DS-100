```python
# lambda functions (anonymous function)

# Lambda function is an anonymous function (nameless) in python.
# it is written by using lambda keywork instead of def.
# Syntax : lambda arguments : expression
# it can contain only one expression and it is linear.
# it automatically return the result of that expression.
# it cannot contains:
    # multiple expression
    # try/except
    # udf 
# Lambda function are mainly used for short and temporary logic.
# Mostly used with HOFs:
    # map()
    # filter()
    # reduce()
```

```python
# HOF (Higher-order-functions).
A Higher-order-function (HOF) is a function that takes another
function and iter(list,tuple,set,string) as arguments , return function or iter.
 # filter(callback_function,iter)
 # map(callback_function,iter)
 # reduce(callback_function,iter,accu)

# Callback.
A callback is a function passed as an argument to another function.
A callback function not executed immediatly it is called later
by the reveiving function.
```

```python
def pass_students(marks):
    pass_stu=[]
    for i in marks:
        if i>=50:
            pass_stu.append(i)
    return pass_stu
marks=[44,5,66,34,45,77,89,31,37,66,42]
res=pass_students(marks)
print(res)
```

```python
def pass_students(marks):
    return marks>=50

marks=[44,5,66,34,45,77,89,31,37,66,42]

res=filter(pass_students,marks)
print(list(res))
```

```python
marks=[44,5,66,34,45,77,89,31,37,66,42]
print(list(filter(lambda n:n>=50,marks)))
```

```python
# wap to filter out elements from a given list.
emp_names=["rohan","aman","kamal","shivani","monika"]
user_char="v"
def filter_name(n):
    new_name=[]
    for i in n:
        if user_char in i:
            new_name.append(i)
    return new_name
res=filter_name(emp_names)
print(res)
```

```python
emp_names=["rohan","aman","kamal","shivani","monika"]
user_char="v"
def filter_name(n):
    return user_char in n
res=filter(filter_name,emp_names)
print(list(res))
```

```python
emp_names=["rohan","aman","kamal","shivani","monika"]
user_char="v"
res=filter(lambda x:user_char in x,emp_names)
print(list(res))
```

```python
my_list=[23,45,67,"43543",5,"Rohan",True,False,"apple",67.90]
def ele_sperate(elm):
    int_list=[]
    str_list=[]
    for i in elm:
        if type(i)==int:
            int_list.append(i)
        if type(i)==str:
            str_list.append(i)
    return [int_list,str_list]
    
res=ele_sperate(my_list)
print(res)
for i in res:
    print(i)
```

```python
my_list=[23,45,67,"aman",5,"Rohan",True,False,"apple",67.90]

def int_sep(elm):
    if type(elm)==int:
        return True

def str_sep(elm):
    if type(elm)==float:
        return True

int_list=filter(int_sep,my_list)
str_list=filter(str_sep,my_list)

print(list(int_list))
print(list(str_list))
```
