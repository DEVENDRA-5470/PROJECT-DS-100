```python
# How to load data in mysql from pandas using connectivity.
import random
import pandas as pd
import mysql.connector
from faker import Faker
fake=Faker()
fake_name=[fake.name() for name in range(4)]
random_salary=[random.randrange(40000,80000) for s in range(4)]
# print(random_salary)
# print(fake_name)

df = pd.DataFrame({
    "id":[i for i in range(100,200)],
    "name":fake_name,
    "salary":random_salary,
})

# create_table="""
#         CREATE TABLE IF NOT EXISTS PANDAS (ID INT , NAME VARCHAR(100), SALARY BIGINT);
#         """
#         agent.execute(create_table)

insert_query="""
        INSERT INTO PANDAS (ID,NAME,SALARY)
        VALUES(%s, %s, %s)
"""

# print(df.to_numpy())
with mysql.connector.connect(
    host="localhost",
    user="root",
    password="root@123",
    database="batch_100"
) as agent:

    with agent.cursor() as cursor:
        data=[tuple(row) for row in df.to_numpy()]
        cursor.executemany(insert_query,data)
        agent.commit()
        print(f"{cursor.rowcount} row inserted..")
df
```
