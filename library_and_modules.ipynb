```python
from datetime import datetime
start=datetime(2010,1,1)
end=datetime(2025,1,1)
cur=start
years=[]
while cur<=end:
    years.append(cur.year)
    cur=cur.replace(year=cur.year+2)
print(years)
```

```python
# write a program to search all number in a string sentence.
# regular expressions(regex) that allows pattern matching , searching
# replace and parsing.

# match : checks only from the beginning of the string.
import re
text="This is 54434 phone number is 78989855555"
pattern=r"\d+"
result=re.match(pattern,text)
print(result)
```

```python
# findall : entire string scan and returns list.
import re
text="This @ # is * ! 54434 phone number is 78989855555"
pattern=r"\w+"
result=re.findall(pattern,text)
print(result)
```

```python
# search : Scans the entire string and returns the first match
# if find anywhere.
import re
text="This is 54434 phone number is 78989855555 878"
pattern=r"\d{11}"
result=re.search(pattern,text)
print(result.group())
```

```python
# fullmatch: requires the entire string to match the pattern.
import re
text="5443478989855555"
pattern=r"\d+"
result=re.fullmatch(pattern,text)
print(result.group())
```
