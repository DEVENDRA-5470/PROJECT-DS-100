```python
API(application programming interface) is a set of rules and protocals which enables
to communicate or interchange infromation between two or more softaware application with
each other in the form of request(client) and response(data).

REST API(Representation state transfer)
REQUEST :HTTP METHOD (GET,POST,PUT,DELETE, etc.)
DATA FORMAT : JSON,XML

TYPES OF APIs:
1.Based on exposure (who can access them).
2.Based on architecture (communication logic design logic)

1.Based on exposure (who can access them).
    1.Public API (Open API) Ex.wheaterapi , twitterapi
    2.Private API (Internal API) Ex.Frontend calls backend (no public access)
    3.Partner API : Access Control through APIs keys , OAuth, IP whitelisting EX.payment gateway.
    4.Composite API : Combines multiple APIs in one request.

2.Based on architecture.
    1.REST API : HTTP METHODS , JSON . EX. web apps , mobile apps ,Public APIs
    2.SOAP API (Simple object access protocol): Only XML , HTTP , SMTP . EX.Banking apis.
    3.GraphQL API.
    4.gRPC(Google Remote Procedure call) : HTTP/2
    5.WebSocket API : full duplex : chats , liver trading system .
```

```python
import requests
import pandas as pd
import mysql.connector
endpoint=r"https://jsonplaceholder.typicode.com/users"
response=requests.get(endpoint)

create_table="""
        CREATE TABLE IF NOT EXISTS APIS_DATA (
        ID INT AUTO_INCREMENT PRIMARY KEY ,
        NAME VARCHAR(100), 
        USERNAME VARCHAR(100),
        EMAIL VARCHAR(200),
        WEBSITE VARCHAR(300),
        CITY VARCHAR(300)
        );
        """
if response.status_code==200:
    data=response.json()
    df=pd.json_normalize(data)

    values=[]
    for i,row in df.iterrows():
        values.append(
            (
             row['name'],
             row["username"],
             row["email"],
             row["website"],
             row["address.city"]
            )
        )
    print(values)

insert="""
INSERT INTO APIS_DATA (NAME,USERNAME,EMAIL,WEBSITE,CITY)
VALUES(%s , %s ,%s ,%s, %s)
"""
       

with mysql.connector.connect(
    host="localhost",
    user="root",
    password="root@123",
    database="batch_100"
) as agent:

    with agent.cursor() as cursor:
        cursor.execute(create_table)
        cursor.executemany(insert,values)
        agent.commit()
        print("Data Inserted..")
       
```
