```python
# Connect
def connect(db=None):
    import mysql.connector as connector
    mydb=connector.connect(
    host="localhost",
    user="root",
    password="root@123",
    database=db
    )
    return mydb


# Create a database using python.
def create_database(db_name=None):
    try:
        conn=connect()
        agent=conn.cursor()
        query=F"CREATE DATABASE {db_name}"
        agent.execute(query)
        print(f"Your {db_name} is Created ✅")
    except Exception as e:
        print("Error:",e)


# Listing of databases
def list_db():
    try:
        conn=connect()
        agent=conn.cursor()
        agent.execute("SHOW DATABASES")
        res=agent.fetchall()
        print(res)
    except Exception as e:
        print("Error:",e)


# Delete databases.
def delete_db(db_list):
    try:
        conn=connect()
        agent=conn.cursor()
        for db in db_list:
            agent.execute(F"DROP DATABASE {db}")
            print(f"Your {db} Delete ✅")
    except Exception as e:
        print("Error:",e)


# CREATE TABLE
def create_table(tb_name=None):
    try:
        conn=connect("batch_100")
        agent=conn.cursor()
        agent.execute(f"SHOW TABLES LIKE '{tb_name}' ")
        result=agent.fetchone()
        
        if result:
            print(f"Table {tb_name} already exit")
            agent.execute("SHOW TABLES")
            for (table,) in agent.fetchall():
                print(table)
        else:
            query=f"CREATE TABLE {tb_name} (stu_name varchar(100) , stu_age int);"
            agent.execute(query)
            conn.commit()
            print(f"Table {tb_name} Created ✅")
            
    except Exception as e:
        print("Error :",e)
    finally:
        conn.close()


# Delete tables.
def delete_tables(tbl_list):
    try:
        conn=connect("batch_100")
        agent=conn.cursor()
        for tb in tbl_list:
            agent.execute(f"DROP TABLE {tb}")
            print(f"{tb} Deleted ✅")
    except Exception as e:
        print("Error :", e)

# change database
def change_db(db_name=None):
    try:
        conn=connect()
        agent=conn.cursor()
        agent.execute(F"USE {db_name}")
    except Exception as e:
        print("Error :", e)


# list current database tables.
def list_tbl(db_name=None):
    try:
        conn=connect(db_name)
        agent=conn.cursor()
        agent.execute("SHOW tables")
        res=agent.fetchall()
        print(res)
    except Exception as e:
        print("Error:",e)
```

```python
while True:
    print("""
============================
1.SHOW LIST OF DATABASES.
2.CREATE NEW DATABASE.
3.CHANGE DATABASE NAME.
4.SHOW LIST OF TABLES.
============================
    """)
    choice=input("ENTER YOU CHOICE : ")
    if choice=="1":
        list_db()
    if choice=="2":
        db_name=input("Enter Your New Database Name :")
        create_database(db_name)
    if choice=="3":
        list_db()
        db=input("Enter DB name :")
        change_db(db)
    if choice=="4":
        list_db()
        db=input("Enter DB name :")
        list_tbl(db)
        
    if choice=="0":
        print("Program closed")
        break;
```
