```python
# Connecty
def connect():
    import mysql.connector as connector
    mydb=connector.connect(
    host="localhost",
    user="root",
    password="root@123"
    )
    return mydb
```

```python
# Create a database using python.
def create_database(db_name):
    try:
        conn=connect()
        agent=conn.cursor()
        query=F"CREATE DATABASE {db_name}"
        agent.execute(query)
        print(f"Your {db_name} is Created ✅")
    except Exception as e:
        print("Error:",e)
create_database("PYTHON_DB")
```

```python
# Listing of databases
def list_db():
    try:
        conn=connect()
        agent=conn.cursor()
        agent.execute("SHOW DATABASES")
        res=agent.fetchall()
        print(res)
    except Exception as e:
        print("Error:",e)
list_db()
```

```python
# Delete databases.
def delete_db(db_list):
    try:
        conn=connect()
        agent=conn.cursor()
        for db in db_list:
            agent.execute(F"DROP DATABASE {db}")
            print(f"Your {db} Delete ✅")
    except Exception as e:
        print("Error:",e)

db_list=['rumi','school','tarun','vish_flask']
delete_db(db_list)
```
