```python
# Connect
def connect(db=None):
    import mysql.connector as connector
    mydb=connector.connect(
    host="localhost",
    user="root",
    password="root@123",
    database=db
    )
    return mydb
```

```python
# Create a database using python.
def create_database(db_name=None):
    try:
        conn=connect()
        agent=conn.cursor()
        query=F"CREATE DATABASE {db_name}"
        agent.execute(query)
        print(f"Your {db_name} is Created ✅")
    except Exception as e:
        print("Error:",e)
create_database("PYTHON_DB")
```

```python
# Listing of databases
def list_db():
    try:
        conn=connect()
        agent=conn.cursor()
        agent.execute("SHOW DATABASES")
        res=agent.fetchall()
        print(res)
    except Exception as e:
        print("Error:",e)
list_db()
```

```python
# Delete databases.
def delete_db(db_list):
    try:
        conn=connect()
        agent=conn.cursor()
        for db in db_list:
            agent.execute(F"DROP DATABASE {db}")
            print(f"Your {db} Delete ✅")
    except Exception as e:
        print("Error:",e)

db_list=['rumi','school','tarun','vish_flask']
delete_db(db_list)
```

```python
# CREATE TABLE
def create_table(tb_name=None):
    try:
        
        conn=connect("batch_100")
        agent=conn.cursor()
        agent.execute(f"SHOW TABLES LIKE '{tb_name}' ")
        result=agent.fetchone()
        
        if result:
            print(f"Table {tb_name} already exit")
            agent.execute("SHOW TABLES")
            for (table,) in agent.fetchall():
                print(table)
        else:
            query=f"CREATE TABLE {tb_name} (stu_name varchar(100) , stu_age int);"
            agent.execute(query)
            conn.commit()
            print(f"Table {tb_name} Created ✅")
            
    except Exception as e:
        print("Error :",e)
    finally:
        conn.close()

create_table("NEW_TABLE")
```

```python
# Delete tables.
def delete_tables(tbl_list):
    try:
        conn=connect("batch_100")
        agent=conn.cursor()
        for tb in tbl_list:
            agent.execute(f"DROP TABLE {tb}")
            print(f"{tb} Deleted ✅")
    except Exception as e:
        print("Error :", e)
tbl_list=["none"]
delete_tables(tbl_list)
```
