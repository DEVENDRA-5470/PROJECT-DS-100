```python
marks=int(input("Enter your Marks : "))
if marks > 50 :
    stu_name=input("Enter you name :")
    if stu_name:
        print(f"Student name is : {stu_name}")
        stu_roll_no=int(input("Enter Your roll : "))
        if stu_roll_no > 100:
            print(f"Student Roll no is : {stu_roll_no}")
            print("You are pass")
        elif stu_roll_no==600:
            print(f"Your Roll no is : {stu_roll_no}")
        elif stu_roll_no==700:
            print(f"Your Roll no is : {stu_roll_no}")
        else:
            print("Roll number is not matching")
    
    else:
        print("Please Provide Student name")
else:
    print("Fail")
```

```python
print("""
Press + for addition
Press - for Sub
press * for multiply
Press / for Divide
""")
option=input("Please Choose an operation from the Menu :")
num1=int(input("Enter Your Number-1 : "))
num2=int(input("Enter Your Number-2 : "))

if option=="+":
    res=num1+num2
    print("Result is :",res)

if option=="-":
    res=num1-num2
    print("Result is :",res)

if option=="*":
    res=num1*num2
    print("Result is :",res)

if option=="/":
    res=num1/num2
    print("Result is :",res)
    
else:
    print("Invalid opertion..")
```
