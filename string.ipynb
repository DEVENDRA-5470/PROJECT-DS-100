```python
import socket

target = "127.0.0.1"
ports = range(1, 1025)

for port in ports:
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.settimeout(0.5)

    result = sock.connect_ex((target, port))
    if result == 0:
        print(f"[OPEN] Port {port}")

    sock.close()
    print("Done")
```
