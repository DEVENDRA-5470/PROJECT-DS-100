```python
# What is file handling.
# File handling is a of creating,opening,reading
# writing ,updating and closing files.
# so a program can store data permanently.

# What is file 
A File is a structured collection of data stroed
on a storage device(SSD,Hard drive etc) with 
a specific name and location managed by OS.

# Types of file.
# Text files (.csv,.docs,.txt)
# Binary file (media files.)

# Path and its types.
# Path is the address of a file or directory in  a file system.
# ex.c:/user/dev/python.py

# 1.Absolute Path
    An absolute path is the full path from the root directory.
    C:\
    D:\
    E:\
# 2.Relative Path
    A Relative path depends on you current working directory (CWD).
    BATCH-100/text.py
    downloads/resume.pdf
    .  current dir
    .. parent dir

# Methods.
# open("file_name","mode",encode=) : Used to open file.
# read()
# readlines()
# write()
# writelines()
# close()
# flush()
# seek()
# tail()


# Modes:
x  : only file creates
w  : file creates and write
w+ : file write and read
r  : file read
r+ : open and read (cursor start from 0)
a  : append data in existing file.

# Binary Modes
rb,wb,ab.

# Operations
# CRUD operation
# Context manager
```

```python
with open("new.txt","w+",encoding="utf-8") as file:
    file.write("This is Done in one time ❌✅ ✅✅✅")
    file.seek(3)
    read=file.read(99)
    print(read)
    print("File done")
```
