```python
# Defination & Properties
# 1.Dictionary is a data structure used to store multiple itmes in the form of
# key:value pair.
# 2.Dictionary are mutable
# 3.Dictionary does not support indexing it works on the keys.
# 4.Dictionary keys must be unique and only support data type(string,num,tuple)
# 5.In dictionary each key:value pair represents one unique entity
# 6.Dictionary formed by using {}.
```

```python
# Creation
my_entity={1:"aman","a":"python",1:"mumbai"}
print(my_entity)
print(len(my_entity))

# method for casting.
dict()
```

```python
my_entity={1:"aman","a":"python"}
print(my_entity)
print(my_entity[1,"a"])
```

```python
# in-built methods.
# key()
my_data={"a":"Delhi",
         "m":"mumbai",
         1:100,
         2:2000,
         "t":"tuesday"}
print(my_data.keys())
```

```python
# values()
my_data={"a":"Delhi",
         "m":"mumbai",
         1:100,
         2:2000,
         "t":"tuesday"}
print(my_data.values())
```

```python
# items()
my_data={"a":"Delhi",
         "m":"mumbai",
         1:100,
         2:2000,
         "t":"tuesday"}
print(my_data.items())
```

```python
# get()
my_data={"a":"Delhi",
         "m":"mumbai",
         1:100,
         2:2000,
         "t":"tuesday"}
print(my_data.get("a","N/A"))
```

```python
# zip()
l1=[1,2,3,4,5]
l2=["a","b"]
res=dict(zip(l2,l1))
print(res)
```

```python
my_data={"a":"Delhi",
         "m":"mumbai",
         1:100,
         2:2000,
         "t":"tuesday"}
res=my_data.pop(1)
print(res)
print(my_data)
```

```python
my_data={
    "b":"noida",
    "a":"Delhi",
         "m":"mumbai",
         1:100,
         2:2000,
         "t":"tuesday"}
my_data.popitem()
print(my_data)
```
